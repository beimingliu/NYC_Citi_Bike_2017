```python
import pandas as pd
import numpy as np
import folium
import pickle
```

```python
PATH = "../data/"
```

```python
!ls {PATH}
```

```python
trips = pd.read_feather(PATH+'nyc_popular_trips')
stations_dict = pickle.load(open( PATH+"stations_dict.p", "rb" ))
stations_latlng = pickle.load(open( PATH+"stations_latlng.p", "rb" ))
```

```python
trips.drop('index',1,inplace=True)
trips.head()
```

```python
stations_dict
```

```python
stations_latlng
```

```python
df = trips.head(2000).copy()
```

```python
df['latlon1'] = df['start station id'].map(stations_latlng)
df['latlon2'] = df['end station id'].map(stations_latlng)
df['log_size'] = np.log(df['size'])
df.head()
```

```python
df['log_size'].quantile([0, .25, .5, .75, 1.])
```

```python
import branca.colormap as cm
linear = cm.LinearColormap(
#     ['#fffee8', '#fff200'],
    ['#007fff','#5fff56','#fff600','#ff0000'],
    vmin=6.70, vmax=8.99,
)

linear.caption = 'log total trips'
linear
```

```python
%%time
NYC = folium.Map(
    location=[40.763940, -73.982651], #lat, longitude
#     tiles='CartoDB positron',
    tiles='CartoDB dark_matter',
    zoom_start=12
)

# len(df)
n = 500
# adjust so stations are visble if pickup and dropoff overlap
adjust = 0.0001
for row in range(n):
    
    # start station
    folium.Circle(
        radius=10,
        location=[df.iloc[row][3][0]-adjust,df.iloc[row][3][1]-adjust],
        color='#56d2ff',
        fill=True,
        fill_color='#56d2ff',
        fill_opacity=.5,
        popup='Pick: '+'<br>'+str(stations_dict[df.iloc[row][0]]),
    ).add_to(NYC)
    
    # end station
    folium.Circle(
        radius=5,
        location=[df.iloc[row][4][0]+adjust,df.iloc[row][4][1]+adjust],
        color='#49ff80',
        fill=True,
        fill_color='#49ff80',
        fill_opacity=.5,
        popup='Dropoff: '+'<br>'+str(stations_dict[df.iloc[row][1]]),
    ).add_to(NYC)
    
    opa = max(np.exp((df.iloc[row][5]-6.704414)/(8.985445-6.704414))-1, .5)
    w = df.iloc[row][5]/8.985445*2
#     print(opa)

    # add line bwt two stations
    folium.PolyLine([[[df.iloc[row][3][0]-adjust,df.iloc[row][3][1]-adjust], 
                      [df.iloc[row][4][0]+adjust,df.iloc[row][4][1]+adjust]]],
                  popup='Trips :'+str(df.iloc[row][2]),
                  color=linear(df.iloc[row][5]),
                  weight=w,
                  opacity= opa
                   ).add_to(NYC)
    
NYC.add_child(linear)
```

```python
NYC.save('NYC%s.html' %n)
```
